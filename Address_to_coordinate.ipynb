# 좌표 -> 위경도

```python
#라이브러리
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('./인천광역시 중구_카페 및 커피숍 현황_20210805.csv', encoding='cp949')
df.head()
```

```python
# 데이터 정제 , 소재지 주소 뒤 N층으로 인한 오류 제거하기 위함
# 데이터프레임 주소값 추출
address = df['소재지도로명주소']
print(address.head())

# 주소 데이터 다듬기
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
print(address)
```

```python
# 도로명 주소 -> 위경도
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent = 'South Korea')

# 위도,경도를 반환하는 함수
def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude,geo.longitude]
    return x_y

# 주소를 위,경도 값으로 반환하기
latitude = []
longitude = []

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
address_df = pd.DataFrame({'카페이름':df['사업장명'],'상세주소':df['소재지도로명주소'],'주소':address,'위도':latitude,'경도':longitude})
address_df.head()
```
